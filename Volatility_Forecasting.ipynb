```python
import yfinance as yf

# List of top 20 stocks by market capitalization on NSE
top_20_stocks = [
    "RELIANCE.NS", "TCS.NS", "HDFCBANK.NS", "HINDUNILVR.NS", "INFY.NS",
    "SBIN.NS", "ICICIBANK.NS", "KOTAKBANK.NS", "BHARTIARTL.NS", "ITC.NS",
    "TATAMOTORS.NS", "ASIANPAINT.NS", "DMART.NS", "HCLTECH.NS", "LT.NS",
    "AXISBANK.NS", "ADANIENT.NS", "SUNPHARMA.NS", "MARUTI.NS", "BAJFINANCE.NS"]

company_data = {}

# Loop through each companu
for stock_symbol in top_20_stocks:
    try:
        data = yf.download(stock_symbol)
        # print(data)
    # Store data in dictionary with company name as key
        company_data[stock_symbol] = data
    except Exception as e:
        print(f"Error fetching data for {stock_symbol}: {e}")

#  data for each company
for stock_symbol, df in company_data.items():
    print(f"Data for {stock_symbol}:")
    print(df.tail())
```

```python
# Function to calculate volatility
def calculate_volatility(returns):
    # Calculate standard deviation of returns
    std_dev = np.std(returns)
    # Annualize standard deviation
    annualized_std_dev = std_dev * np.sqrt(252)
    return annualized_std_dev
```

```python
# Loop through each company's data
for stock_symbol, df in company_data.items():
    # Calculate daily returns
    df['Returns'] = df['Adj Close'].pct_change()
    returns = df['Returns'].dropna()

    # Calculate historical volatility
    volatility = calculate_volatility(returns)

    # Plot adjusted closing prices and volatility
    fig, ax1 = plt.subplots()

    # Plot stock prices
    color = 'tab:blue'
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Stock Price', color=color)
    ax1.plot(df.index, df['Adj Close'], color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    # Create a second y-axis for volatility
    ax2 = ax1.twinx()
    color = 'tab:red'
    ax2.set_ylabel('Historical Volatility', color=color)
    ax2.plot(returns.index, returns.rolling(window=30).std() * np.sqrt(252), color=color)  # rolling 30-day volatility
    # ax2.axhline(volatility, color='black', linestyle='--')  # historical volatility as horizontal line
    ax2.tick_params(axis='y', labelcolor=color)

    # Title and legend
    plt.title(f'Stock Prices and Volatility for {stock_symbol}')
    fig.tight_layout()
    plt.show()
```

```python
import os

# Create a folder named "volatility_data" to sotre the forecasted and actual volatility
folder_name = "volatility_data"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Define start and end dates for forecasting
start_date = '2018-08-01'
end_date = '2023-08-31'

# Loop through each company's data
for stock_symbol, df in company_data.items():
    try:
        # Filter data for the specified forecasting period
        df_forecast = df.loc[start_date:end_date].copy()
        y_rescaled = df_forecast['Adj Close'] * 100

        df_forecast['Returns'] = y_rescaled.pct_change()
        returns = df_forecast['Returns'].dropna()

        garch_model = arch_model(returns, vol='Garch', p=1, q=1, rescale=False)
        garch_model_fit = garch_model.fit(disp='off')

        forecasts = garch_model_fit.forecast(start=returns.index[0])

        # Extract values as one-dimensional arrays
        actual_volatility = np.sqrt(252) * returns.rolling(window=30).std().values
        forecasted_volatility = (15 * np.sqrt(forecasts.variance.dropna())).values

        # Create DataFrame for actual volatility and forecasted volatility
        volatility_data = pd.DataFrame({
            'Date': returns.index,
            'Actual Volatility': actual_volatility.flatten(),  # Ensure one-dimensional array
            'Forecasted Volatility': forecasted_volatility.flatten()  # Ensure one-dimensional array
        })

        # Save DataFrame to Excel file in the folder
        excel_file_name = os.path.join(folder_name, f"{stock_symbol}_volatility_forecasts.xlsx")
        volatility_data.to_excel(excel_file_name, index=False)

        print(f"Successfully saved volatility forecasts for {stock_symbol} to {excel_file_name}")

        # Plot actual volatility and forecasted volatility
        plt.figure(figsize=(10, 6))
        plt.plot(returns.index, np.sqrt(252) * returns.rolling(window=30).std(), label='Actual Volatility (30-day rolling)')
        plt.plot(forecasts.variance.dropna().index, np.sqrt(forecasts.variance.dropna())*15, label='Forecasted Volatility')
        plt.title(f'Volatility Forecasting using GARCH(1,1) for {stock_symbol}')
        plt.xlabel('Date')
        plt.ylabel('Volatility')
        plt.legend()
        plt.show()

    except Exception as e:
        print(f"Error processing volatility forecasts for {stock_symbol}: {e}")
```

```python
# Define function to determine market condition
def determine_market_condition(price_changes):
    if all(change > 0 for change in price_changes):
        return 'Bullish'
    elif all(change < 0 for change in price_changes):
        return 'Bearish'
    else:
        return 'Neutral'

# # Define function to determine market condition
# def determine_market_condition(price_changes):
#     bullish_days = 0
#     bearish_days = 0

#     for change in price_changes:
#         if change >= 0:  # Allowing for tolerance in price decreases
#             bullish_days += 1
#             bearish_days = 0
#         elif -0.01 <= change < 0:  # Allowing for a slight price decrease within tolerance
#             bullish_days += 1
#             bearish_days = 0
#         elif change < 0:  # Allowing for tolerance in price increases
#             bearish_days += 1
#             bullish_days = 0

#         if bullish_days == 5:
#             return 'Bullish'
#         elif bearish_days == 5:
#             return 'Bearish'

#     return 'Neutral'

 # Example:
# price_changes = [0.01, 0.02, -0.005, 0.005, -0.02, 0.01, 0.005]
# market_condition = determine_market_condition(price_changes)
# print("Market Condition:", market_condition)

# Define function to implement trading strategy
def implement_trading_strategy(volatility_data):
    trades = []
    for i in range(len(volatility_data)):
        current_volatility = np.sqrt(252) * volatility_data.iloc[i, 2]
        actual_volatility = volatility_data.iloc[i, 1]
        price_changes = volatility_data.iloc[max(0, i-4):i+1, 0].diff().dropna().tolist()
        market_condition = determine_market_condition(price_changes)

        # Trading signal based on volatility comparison and market condition
        if current_volatility > actual_volatility:
            if market_condition == 'Bullish':
                trades.append('Buy')
            elif market_condition == 'Bearish':
                trades.append('Sell')
            else:
                trades.append('Neutral')
        else:
            trades.append('Neutral')

    return trades

# Implement trading strategy for each stock
for stock_symbol, df in company_data.items():
    try:
        # Filter data for the specified forecasting period
        df_forecast = df.loc[start_date:end_date].copy()
        y_rescaled = df_forecast['Adj Close'] * 100
        df_forecast['Returns'] = y_rescaled.pct_change()

        returns = df_forecast['Returns'].dropna()

        garch_model = arch_model(returns, vol='Garch', p=1, q=1, rescale=False)
        garch_model_fit = garch_model.fit(disp='off')

        # Forecast volatility
        forecasts = garch_model_fit.forecast(start=returns.index[0])

        # Store forecasted volatility in DataFrame
        df_forecast['Forecasted Volatility'] = np.sqrt(252) * forecasts.variance.dropna()

        # trading strategy
        trades = implement_trading_strategy(df_forecast)

        # Add trading signals to the DataFrame
        df_forecast['Trading Signal'] = trades

        # DataFrame with trading signals
        print(f"Trading signals for {stock_symbol}:")
        print(df_forecast[['Adj Close', 'Forecasted Volatility', 'Trading Signal']].head())
    except Exception as e:
        print(f"Error implementing trading strategy for {stock_symbol}: {e}")
```

```python
# Calculate returns based on buy and sell signals
def calculate_returns(df):
    returns = []
    position = None
    for i in range(1, len(df)):
        if df.iloc[i-1]['Trading Signal'] == 'Buy' and position != 'Buy':
            buy_price = df.iloc[i]['Adj Close']
            position = 'Buy'
        elif df.iloc[i-1]['Trading Signal'] == 'Sell' and position == 'Buy':
            sell_price = df.iloc[i]['Adj Close']
            returns.append((sell_price - buy_price) / buy_price)  #  return percentage
            position = None
    return returns

# Calculate returns for each stock
returns_per_stock = {}
for stock_symbol, df in company_data.items():
    try:
        df_forecast = df.loc[start_date:end_date].copy()
        trades = implement_trading_strategy(df_forecast)
        df_forecast['Trading Signal'] = trades
        returns = calculate_returns(df_forecast)
        returns_per_stock[stock_symbol] = returns
    except Exception as e:
        print(f"Error calculating returns for {stock_symbol}: {e}")

# Print average returns for each stock
for stock_symbol, returns in returns_per_stock.items():
    if len(returns) > 0:
        avg_return = sum(returns) / len(returns)
        print(f"Average return for {stock_symbol} as per our trading strategy: {avg_return:.2%}")
    else:
        print(f"No trades executed for {stock_symbol}")
```

```python
# Calculate returns based on buy and sell signals with trade details
def calculate_returns_with_details(df):
    trades = []
    position = None
    for i in range(1, len(df)):
        if df.iloc[i-1]['Trading Signal'] == 'Buy' and position != 'Buy':
            buy_date = df.index[i]
            buy_price = df.iloc[i]['Adj Close']
            position = 'Buy'
        elif df.iloc[i-1]['Trading Signal'] == 'Sell' and position == 'Buy':
            sell_date = df.index[i]
            sell_price = df.iloc[i]['Adj Close']
            returns = (sell_price - buy_price) / buy_price
            trade_details = {'Buy Date': buy_date, 'Buy Price': buy_price, 'Sell Date': sell_date, 'Sell Price': sell_price, 'Return': returns}
            trades.append(trade_details)
            position = None
    return trades

# Calculate returns with trade details for each stock
returns_details_per_stock = {}
for stock_symbol, df in company_data.items():
    try:
        df_forecast = df.loc[start_date:end_date].copy()
        trades = implement_trading_strategy(df_forecast)
        df_forecast['Trading Signal'] = trades
        trades_details = calculate_returns_with_details(df_forecast)
        returns_details_per_stock[stock_symbol] = trades_details
    except Exception as e:
        print(f"Error calculating returns with details for {stock_symbol}: {e}")

# Print trade details for each stock
for stock_symbol, trades_details in returns_details_per_stock.items():
    print(f"Trade details for {stock_symbol}:")
    for i, trade in enumerate(trades_details):
        print(f"Trade {i+1}:")
        print(f"  Buy Date: {trade['Buy Date']}, Buy Price: {trade['Buy Price']:.2f}")
        print(f"  Sell Date: {trade['Sell Date']}, Sell Price: {trade['Sell Price']:.2f}")
        print(f"  Return: {trade['Return']:.2%}")
    print()
```

```python
import numpy as np

# Calculate Sharpe ratio
def calculate_sharpe_ratio(returns):
    mean_return = np.mean(returns)
    std_dev = np.std(returns)
    sharpe_ratio = mean_return / std_dev
    return sharpe_ratio

# Calculate maximum drawdown
def calculate_max_drawdown(returns):
    cumulative_returns = np.cumprod(1 + np.array(returns))
    peak = np.maximum.accumulate(cumulative_returns)
    drawdown = (cumulative_returns - peak) / peak
    max_drawdown = np.min(drawdown)
    return max_drawdown

# # Calculate turnover
# def calculate_turnover(trades_details):
#     buy_dates = [trade['Buy Date'] for trade in trades_details]
#     sell_dates = [trade['Sell Date'] for trade in trades_details]
#     total_trading_days = len(np.unique(buy_dates + sell_dates))
#     total_trades = len(trades_details)
#     turnover = total_trades / total_trading_days
#     return turnover

# Evaluate risk management for each stock
risk_metrics_per_stock = {}
for stock_symbol, trades_details in returns_details_per_stock.items():
    try:
        returns = [trade['Return'] for trade in trades_details]
        sharpe_ratio = calculate_sharpe_ratio(returns)
        max_drawdown = calculate_max_drawdown(returns)

        # # Calculate turnover
        # turnover = calculate_turnover(trades_details)

        # Store risk metrics in dictionary
        risk_metrics_per_stock[stock_symbol] = {'Sharpe Ratio': sharpe_ratio, 'Max Drawdown': max_drawdown}
    except Exception as e:
        print(f"Error calculating risk metrics for {stock_symbol}: {e}")

# Print risk metrics for each stock
for stock_symbol, risk_metrics in risk_metrics_per_stock.items():
    print(f"Risk metrics for {stock_symbol}:")
    for metric, value in risk_metrics.items():
        print(f"  {metric}: {value:.4f}")
    print()
```
